```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('cars_dataset.csv')
```

```python
df
```

```python
df['owner'].value_counts()
```

```python
df['brand'].unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                                                    df.drop(columns=['selling_price']),
                                                    df['selling_price'], 
                                                    test_size=0.2, 
                                                    random_state=42
                                                )
    
```

```python
# Applying Ordinal Encoder on Owner
oe = OrdinalEncoder(categories=[['Test Drive Car', 'Fourth & Above Owner', 'Third Owner', 'Second Owner', 'First Owner']])

X_train_owner = oe.fit_transform(X_train.loc[:, ['owner']])
X_test_owner = oe.transform(X_test.loc[:, ['owner']])
```

```python
# Jab OrdinalEncoder encode karta hai, result aata hai as a (NumPy Array):
# Convert array to df 
X_train_owner_df = pd.DataFrame(X_train_owner, columns=oe.get_feature_names_out())
X_test_owner_df = pd.DataFrame(X_test_owner, columns=oe.get_feature_names_out())
```

```python
X_train_owner_df.head()
```

```python
# Applying the OHE on 'brand' and 'fuel'  
ohe = OneHotEncoder(sparse_output=False)

X_train_brand_fuel = ohe.fit_transform(X_train[['brand', 'fuel']])
X_test_brand_fuel = ohe.transform(X_test[['brand', 'fuel']])
```

```python
# Converting into dataframe
X_train_brand_fuel_df = pd.DataFrame(X_train_brand_fuel, columns=ohe.get_feature_names_out())
X_test_brand_fuel_df = pd.DataFrame(X_test_brand_fuel, columns=ohe.get_feature_names_out())
```

```python
print(X_train_brand_fuel.shape)          # how many cols in data?
print(len(ohe.get_feature_names_out()))  # how many col names?
```

```python
X_train_brand_fuel_df.head()
```

```python
X_train.head()
```

```python
# To delete the column of 'brand, 'fuel' and 'owner' from original df
X_train_rem = X_train.drop(columns=['brand', 'fuel', 'owner'], inplace=True)
X_test_rem = X_test.drop(columns=['brand', 'fuel', 'owner'], inplace=True)
```

```python
X_train_rem      
```

```python
# concat the Encoded col to original df
X_train = pd.concat([X_train_rem, X_train_owner_df, X_train_brand_fuel_df], axis=1)
X_test = pd.concat([X_test_rem, X_test_owner_df, X_test_brand_fuel_df], axis=1)
```

## The Easy Way!

```python
# This process is very hectic to perfrom on big col data, to replace this, Here's comes "Column Transfromer" its very helpful to perfrom different Encoding on column at a time and at the end "marge" them
```

```python
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('cars_dataset.csv')
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                                                      df.drop(columns=['selling_price']),
                                                      df['selling_price'],
                                                      test_size=0.2,
                                                      random_state=42
                                                   )
```

```python
X_train.head()
```

```python
# ColumnTransformer(
#       [
#           ("name1", object/encoding[col1, col2]),
#           ( ),
#           ( )
#       ]
# )

# this the flow to create ColumnTransformer
```

```python
transformers = ColumnTransformer(
      [
            ("ordinal", OrdinalEncoder(categories=[['Test Drive Car', 'Fourth & Above Owner', 'Third Owner', 'Second Owner', 'First Owner']]),['owner']),
            ("onehot", OneHotEncoder(sparse_output=False), ['brand', 'fuel'])
      ],  
      remainder='passthrough'  #it means ki baki col mein kuch ni karna, as it is pass hone do  
)

# setting to get a direct pandas df (quick method)
transformers.set_output(transform='pandas')
```

```python
X_train_transformed = transformers.fit_transform(X_train)
X_test_transformed = transformers.transform(X_train)
```
